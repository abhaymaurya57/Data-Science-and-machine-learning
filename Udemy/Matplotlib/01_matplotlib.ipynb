```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,10)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x,y,'r-')
# plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('x Lable')
plt.ylabel('y Lable')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.title("First Plot")

plt.subplot(1,2,2)
plt.plot(x,y,'b')
plt.title("Second Plot")
```

```python
#  OO
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Lable')
axes.set_ylabel('Y Lable')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOAT')

axes2.plot(x,y)
axes2.set_title('SMALLER PLOAT')
```

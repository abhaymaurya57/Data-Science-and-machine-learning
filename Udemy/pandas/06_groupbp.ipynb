```python
import numpy as np
import pandas as pd
```

```python
data={'Company':['GOOG','GOOG','MSEF','MSEF','FB','FB'],
      'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
      'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
byComp=df.groupby('Company')
```

```python
byComp
```

```python
byComp['Sales'].mean()
```

```python
byComp.mean(numeric_only=True)
```

```python
byComp.sum()
```

```python
byComp.std(numeric_only=True)
```

```python
byComp.sum().loc[['FB','GOOG']]
```

```python
byComp.sum().loc['FB']
```
